```python
import os
import json
from pathlib import Path

# === CONFIGURATION ===
dataset_path = '/content/drive/MyDrive/ExperimentalDataset (4)/ExperimentalDataset'  # <-- Update if needed
output_dir = Path('/content/drive/MyDrive/ExperimentalDataset_phase1_output')
output_dir.mkdir(parents=True, exist_ok=True)
components_list = []

def extract_components_from_node(node, app_name, screen_name, parent_id='', depth=0):
    extracted = []
    node_class = node.get('class')
    bounds = node.get('bounds')
    # Only extract if valid class and bounds (list of 4 numbers)
    if node_class and isinstance(bounds, list) and len(bounds) == 4 and all(isinstance(x, (int, float)) for x in bounds):
        comp_id = f"{app_name}_{screen_name}_{depth}_{len(extracted)}"
        component = {
            'id': comp_id,
            'app': app_name,
            'screen': screen_name,
            'class': node_class,
            'bounds': bounds,
            'depth': depth,
            'parent_id': parent_id
        }
        extracted.append(component)
    # Recursively extract children
    for child in node.get('children', []):
        extracted.extend(extract_components_from_node(child, app_name, screen_name, parent_id=comp_id, depth=depth+1))
    return extracted

# === MAIN EXTRACTION LOOP ===
for app_dir in Path(dataset_path).iterdir():
    if not app_dir.is_dir():
        continue
    app_name = app_dir.name
    print(f"Processing app: {app_name}")
    for json_file in app_dir.glob("*.json"):
        screen_name = json_file.stem
        try:
            with open(json_file, 'r') as f:
                data = json.load(f)
            # Defensive: skip empty or list-rooted JSONs
            if not isinstance(data, dict):
                print(f"  Skipping (not a dict): {json_file}")
                continue
            if 'activity' not in data or 'root' not in data['activity']:
                print(f"  Skipping (no activity/root): {json_file}")
                continue
            root = data['activity']['root']
            comps = extract_components_from_node(root, app_name, screen_name)
            components_list.extend(comps)
            print(f"  {screen_name}: {len(comps)} components extracted")
        except Exception as e:
            print(f"Error processing {json_file}: {e}")

print(f"Total components extracted: {len(components_list)}")

# === SAVE ALL COMPONENT METADATA ===
with open(output_dir / "manually_processed_components.json", "w") as f:
    json.dump(components_list, f, indent=2)
print(f"Saved all component metadata to {output_dir}/manually_processed_components.json")
```

```python
import json
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import torch.optim as optim
from pathlib import Path
import os
from collections import defaultdict
from tqdm import tqdm

# === CONFIGURATION ===
dataset_path = '/content/drive/MyDrive/ExperimentalDataset_phase1_output'
checkpoint_dir = Path(dataset_path) / 'checkpoints'
checkpoint_dir.mkdir(parents=True, exist_ok=True)

BATCH_SIZE = 128  # Adjust as per your RAM
EPOCHS = 40
G_LR = 1e-3
D_LR = 1e-4
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
MAX_SEQ_LENGTH = 100
D_STEPS = 1
G_STEPS = 3
NUM_WORKERS = 2  # Per system warning

print(f"Using device: {DEVICE}")

# === DATA PREPARATION FUNCTIONS ===
def create_training_sequences(components):
    screen_components = defaultdict(list)
    for comp in components:
        key = f"{comp['app']}_{comp['screen']}"
        screen_components[key].append(comp)
    training_sequences = []
    for screen_key, comps in screen_components.items():
        if len(comps) > 0:
            sorted_comps = sorted(comps, key=lambda x: x['depth'])
            sequence = {
                'app': sorted_comps[0]['app'],
                'screen': sorted_comps[0]['screen'],
                'sequence': [{'class': comp['class'], 'bounds': comp['bounds']} for comp in sorted_comps]
            }
            training_sequences.append(sequence)
    with open(Path(dataset_path) / 'training_sequences.json', 'w') as f:
        json.dump(training_sequences, f, indent=2)
    print(f"Created {len(training_sequences)} training sequences")
    return training_sequences

def build_vocabulary(components):
    vocab = {'<PAD>': 0, '<START>': 1, '<END>': 2, '<UNK>': 3}
    for comp in components:
        cname = comp['class']
        if cname not in vocab:
            vocab[cname] = len(vocab)
    with open(Path(dataset_path) / 'vocab.json', 'w') as f:
        json.dump(vocab, f, indent=2)
    print(f"Vocabulary size: {len(vocab)}")
    return vocab

# === DATASET CLASS ===
class GUISequenceDataset(Dataset):
    def __init__(self, sequences, vocab, max_length=MAX_SEQ_LENGTH):
        self.vocab = vocab
        self.max_length = max_length
        self.data = []
        for seq in sequences:
            tokens = [vocab['<START>']]
            for comp in seq['sequence']:
                tokens.append(vocab.get(comp['class'], vocab['<UNK>']))
            tokens.append(vocab['<END>'])
            if len(tokens) > max_length:
                tokens = tokens[:max_length]
            else:
                tokens.extend([vocab['<PAD>']] * (max_length - len(tokens)))
            self.data.append(torch.tensor(tokens, dtype=torch.long))
    def __len__(self):
        return len(self.data)
    def __getitem__(self, idx):
        return self.data[idx]

# === MODEL DEFINITIONS ===
class GUIGenerator(nn.Module):
    def __init__(self, vocab_size, embed_dim=256, hidden_dim=512, num_layers=2):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.lstm = nn.LSTM(embed_dim, hidden_dim, num_layers, batch_first=True)
        self.output_projection = nn.Linear(hidden_dim, vocab_size)
        self.dropout = nn.Dropout(0.1)
    def forward(self, x, hidden=None):
        embedded = self.embedding(x)
        embedded = self.dropout(embedded)
        output, hidden = self.lstm(embedded, hidden)
        logits = self.output_projection(output)
        return logits, hidden
    def sample(self, start_token, max_length, device, vocab):
        self.eval()
        with torch.no_grad():
            sequence = [start_token]
            hidden = None
            input_token = torch.tensor([[start_token]], device=device)
            for _ in range(max_length - 1):
                logits, hidden = self.forward(input_token, hidden)
                probs = F.softmax(logits[0, -1], dim=-1)
                next_token = torch.multinomial(probs, 1).item()
                sequence.append(next_token)
                if next_token == vocab['<END>']:
                    break
                input_token = torch.tensor([[next_token]], device=device)
        self.train()
        return sequence

class GUIDiscriminator(nn.Module):
    def __init__(self, vocab_size, embed_dim=256, hidden_dim=256):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.lstm = nn.LSTM(embed_dim, hidden_dim, batch_first=True, bidirectional=True)
        self.classifier = nn.Linear(hidden_dim * 2, 1)
        self.dropout = nn.Dropout(0.2)
    def forward(self, x):
        embedded = self.embedding(x)
        embedded = embedded + 0.05 * torch.randn_like(embedded)
        _, (hidden, _) = self.lstm(embedded)
        features = torch.cat([hidden[-2], hidden[-1]], dim=1)
        output = torch.sigmoid(self.classifier(self.dropout(features)))
        return output

# === CHECKPOINT FUNCTIONS ===
def save_checkpoint(epoch, generator, discriminator, g_optimizer, d_optimizer, vocab, loss_history):
    checkpoint = {
        'epoch': epoch,
        'generator_state_dict': generator.state_dict(),
        'discriminator_state_dict': discriminator.state_dict(),
        'g_optimizer_state_dict': g_optimizer.state_dict(),
        'd_optimizer_state_dict': d_optimizer.state_dict(),
        'vocab': vocab,
        'loss_history': loss_history
    }
    checkpoint_path = checkpoint_dir / f'checkpoint_epoch_{epoch}.pt'
    torch.save(checkpoint, checkpoint_path)
    checkpoints = sorted(checkpoint_dir.glob('checkpoint_epoch_*.pt'))
    if len(checkpoints) > 5:
        for old_checkpoint in checkpoints[:-5]:
            os.remove(old_checkpoint)
    print(f"✅ Checkpoint saved: {checkpoint_path}")
    return checkpoint_path

def load_latest_checkpoint(generator, discriminator, g_optimizer, d_optimizer):
    checkpoints = sorted(checkpoint_dir.glob('checkpoint_epoch_*.pt'))
    if not checkpoints:
        print("No checkpoints found. Starting from scratch.")
        return 0, [], None
    latest_checkpoint = checkpoints[-1]
    print(f"🔄 Loading checkpoint: {latest_checkpoint}")
    checkpoint = torch.load(latest_checkpoint, map_location=DEVICE)
    generator.load_state_dict(checkpoint['generator_state_dict'])
    discriminator.load_state_dict(checkpoint['discriminator_state_dict'])
    g_optimizer.load_state_dict(checkpoint['g_optimizer_state_dict'])
    d_optimizer.load_state_dict(checkpoint['d_optimizer_state_dict'])
    for state in g_optimizer.state.values():
        for k, v in state.items():
            if isinstance(v, torch.Tensor):
                state[k] = v.to(DEVICE)
    for state in d_optimizer.state.values():
        for k, v in state.items():
            if isinstance(v, torch.Tensor):
                state[k] = v.to(DEVICE)
    start_epoch = checkpoint['epoch'] + 1
    loss_history = checkpoint.get('loss_history', [])
    vocab = checkpoint['vocab']
    print(f"✅ Resumed from epoch {checkpoint['epoch']}")
    return start_epoch, loss_history, vocab

# === MAIN TRAINING FUNCTION ===
def train_gui_gan():
    try:
        with open(Path(dataset_path) / 'training_sequences.json', 'r') as f:
            training_sequences = json.load(f)
        with open(Path(dataset_path) / 'vocab.json', 'r') as f:
            vocab = json.load(f)
        print("Loaded existing sequences and vocabulary")
    except FileNotFoundError:
        print("Creating training sequences and vocabulary...")
        with open(Path(dataset_path) / 'manually_processed_components.json', 'r') as f:
            components = json.load(f)
        training_sequences = create_training_sequences(components)
        vocab = build_vocabulary(components)
    inv_vocab = {v: k for k, v in vocab.items()}
    vocab_size = len(vocab)
    dataset = GUISequenceDataset(training_sequences, vocab)
    dataloader = DataLoader(
        dataset,
        batch_size=BATCH_SIZE,
        shuffle=True,
        num_workers=NUM_WORKERS,
        pin_memory=True
    )
    print(f"Dataset size: {len(dataset)} sequences")
    print(f"Vocabulary size: {vocab_size}")
    generator = GUIGenerator(vocab_size).to(DEVICE)
    discriminator = GUIDiscriminator(vocab_size).to(DEVICE)
    g_optimizer = optim.Adam(generator.parameters(), lr=G_LR)
    d_optimizer = optim.Adam(discriminator.parameters(), lr=D_LR)
    start_epoch, loss_history, checkpoint_vocab = load_latest_checkpoint(
        generator, discriminator, g_optimizer, d_optimizer
    )
    if checkpoint_vocab:
        vocab = checkpoint_vocab
        inv_vocab = {v: k for k, v in vocab.items()}
    print(f"🚀 Starting training from epoch {start_epoch}")
    try:
        for epoch in range(start_epoch, EPOCHS):
            epoch_d_losses = []
            epoch_g_losses = []
            # tqdm progress bar for batches
            for batch_idx, real_sequences in enumerate(tqdm(dataloader, desc=f"Epoch {epoch+1}")):
                real_sequences = real_sequences.to(DEVICE)
                batch_size = real_sequences.size(0)
                # --- Discriminator ---
                for _ in range(D_STEPS):
                    d_optimizer.zero_grad()
                    fake_sequences = []
                    for _ in range(batch_size):
                        fake_seq = generator.sample(vocab['<START>'], MAX_SEQ_LENGTH, DEVICE, vocab)
                        fake_seq += [vocab['<PAD>']] * (MAX_SEQ_LENGTH - len(fake_seq))
                        fake_sequences.append(fake_seq[:MAX_SEQ_LENGTH])
                    fake_sequences = torch.tensor(fake_sequences, device=DEVICE)
                    real_pred = discriminator(real_sequences)
                    fake_pred = discriminator(fake_sequences)
                    real_labels = torch.ones_like(real_pred) * 0.9
                    fake_labels = torch.zeros_like(fake_pred) + 0.1
                    d_loss = F.binary_cross_entropy(real_pred, real_labels) + \
                             F.binary_cross_entropy(fake_pred, fake_labels)
                    d_loss.backward()
                    d_optimizer.step()
                    epoch_d_losses.append(d_loss.item())
                # --- Generator ---
                for _ in range(G_STEPS):
                    g_optimizer.zero_grad()
                    fake_sequences = []
                    for _ in range(batch_size):
                        fake_seq = generator.sample(vocab['<START>'], MAX_SEQ_LENGTH, DEVICE, vocab)
                        fake_seq += [vocab['<PAD>']] * (MAX_SEQ_LENGTH - len(fake_seq))
                        fake_sequences.append(fake_seq[:MAX_SEQ_LENGTH])
                    fake_sequences = torch.tensor(fake_sequences, device=DEVICE)
                    fake_pred = discriminator(fake_sequences)
                    g_loss = F.binary_cross_entropy(fake_pred, torch.ones_like(fake_pred))
                    g_loss.backward()
                    g_optimizer.step()
                    epoch_g_losses.append(g_loss.item())
                # DEBUG: Print after every 5 batches
                if (batch_idx+1) % 5 == 0:
                    print(f"Epoch {epoch+1}, Batch {batch_idx+1}/{len(dataloader)}: D Loss={d_loss.item():.4f}, G Loss={g_loss.item():.4f}")
            avg_d_loss = sum(epoch_d_losses) / len(epoch_d_losses) if epoch_d_losses else 0
            avg_g_loss = sum(epoch_g_losses) / len(epoch_g_losses) if epoch_g_losses else 0
            loss_history.append({'epoch': epoch, 'd_loss': avg_d_loss, 'g_loss': avg_g_loss})
            print(f"Epoch {epoch+1}/{EPOCHS} - D Loss: {avg_d_loss:.4f}, G Loss: {avg_g_loss:.4f}")
            if (epoch + 1) % 5 == 0:
                sample_seq = generator.sample(vocab['<START>'], 20, DEVICE, vocab)
                sample_classes = [inv_vocab.get(token, '<UNK>') for token in sample_seq]
                print(f"Sample generated: {sample_classes[:10]}")
            # === SAVE CHECKPOINT EVERY EPOCH ===
            save_checkpoint(epoch, generator, discriminator, g_optimizer, d_optimizer, vocab, loss_history)
        final_checkpoint_path = save_checkpoint(EPOCHS-1, generator, discriminator, g_optimizer, d_optimizer, vocab, loss_history)
        print(f"🎉 Training completed! Final model saved at: {final_checkpoint_path}")
    except KeyboardInterrupt:
        print("\n⚠️ Training interrupted by user")
        save_checkpoint(epoch, generator, discriminator, g_optimizer, d_optimizer, vocab, loss_history)
        print("🚨 Emergency checkpoint saved!")
    except Exception as e:
        print(f"❌ Training error: {e}")
        try:
            save_checkpoint(epoch, generator, discriminator, g_optimizer, d_optimizer, vocab, loss_history)
            print("🚨 Emergency checkpoint saved!")
        except:
            print("❌ Could not save emergency checkpoint")

if __name__ == "__main__":
    train_gui_gan()
```

```python
import os
import json
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from pathlib import Path
from tqdm import tqdm
import shutil

# === CONFIGURATION ===
PHASE1_DATA_PATH = '/content/drive/MyDrive/ExperimentalDataset_phase1_output'
PHASE2_DATA_PATH = '/content/drive/MyDrive/ExperimentalDataset_phase2'
CHECKPOINT_DIR = Path(PHASE2_DATA_PATH) / 'checkpoints'
CHECKPOINT_DIR.mkdir(parents=True, exist_ok=True)
SAMPLE_OUTPUT_PATH = '/content/drive/MyDrive/ExperimentalDataset_phase2/phase2_samples.jsonl'

BATCH_SIZE = 64
EPOCHS = 45
MAX_COMPONENTS = 25
G_LR = 2e-3
D_LR = 5e-5
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
NUM_WORKERS = 0
PIN_MEMORY = True
D_STEPS = 1
G_STEPS = 5
PRINT_FREQ = 20

print(f"Using device: {DEVICE}")

# === DATA PREPARATION ===
def prepare_phase2_data():
    os.makedirs(PHASE2_DATA_PATH, exist_ok=True)
    shutil.copy(
        os.path.join(PHASE1_DATA_PATH, 'vocab.json'),
        os.path.join(PHASE2_DATA_PATH, 'vocab.json')
    )
    with open(os.path.join(PHASE1_DATA_PATH, 'manually_processed_components.json')) as f:
        components = json.load(f)
    gui_metadata = []
    screen_components = {}
    for comp in components:
        screen_key = f"{comp['app']}_{comp['screen']}"
        if screen_key not in screen_components:
            screen_components[screen_key] = []
        screen_components[screen_key].append(comp)
    for screen_key, comps in screen_components.items():
        if len(comps) > 0:
            sorted_comps = sorted(comps, key=lambda x: x.get('depth', 0))
            gui_data = {
                'app': sorted_comps[0]['app'],
                'screen': sorted_comps[0]['screen'],
                'components': []
            }
            for comp in sorted_comps:
                gui_data['components'].append({
                    'type': comp['class'],
                    'bounds': comp.get('bounds', [0, 0, 100, 100])
                })
            gui_metadata.append(gui_data)
    with open(os.path.join(PHASE2_DATA_PATH, 'gui_metadata.json'), 'w') as f:
        json.dump(gui_metadata, f, indent=2)
    print(f"Created {len(gui_metadata)} GUI layouts for Phase 2")
    return gui_metadata

# === DATASET CLASS ===
class GUILayoutDataset(Dataset):
    def __init__(self, metadata_path, vocab_path):
        with open(metadata_path) as f:
            self.data = json.load(f)
        with open(vocab_path) as f:
            self.vocab = json.load(f)
        self.inv_vocab = {v: k for k, v in self.vocab.items()}
        self.sequences = []
        self.layouts = []
        for gui in self.data:
            seq = [self.vocab['<START>']]
            layout = [[0, 0, 0, 0]]
            for c in gui['components'][:MAX_COMPONENTS-2]:
                token = self.vocab.get(c['type'], self.vocab['<UNK>'])
                seq.append(token)
                bounds = c.get('bounds', [0, 0, 100, 100])
                normalized_bounds = [
                    bounds[0] / 1000.0,
                    bounds[1] / 1000.0,
                    bounds[2] / 1000.0,
                    bounds[3] / 1000.0
                ]
                layout.append(normalized_bounds)
            seq.append(self.vocab['<END>'])
            layout.append([0, 0, 0, 0])
            while len(seq) < MAX_COMPONENTS:
                seq.append(self.vocab['<PAD>'])
                layout.append([0, 0, 0, 0])
            self.sequences.append(seq[:MAX_COMPONENTS])
            self.layouts.append(layout[:MAX_COMPONENTS])
    def __len__(self):
        return len(self.sequences)
    def __getitem__(self, idx):
        return {
            'sequence': torch.tensor(self.sequences[idx], dtype=torch.long),
            'layout': torch.tensor(self.layouts[idx], dtype=torch.float)
        }

# === MODEL ARCHITECTURES ===
class LayoutGenerator(nn.Module):
    def __init__(self, vocab_size, embed_dim=256, hidden_dim=512, num_layers=2):
        super().__init__()
        self.vocab_size = vocab_size
        self.embed = nn.Embedding(vocab_size, embed_dim)
        self.lstm = nn.LSTM(embed_dim, hidden_dim, num_layers, batch_first=True)
        self.fc_token = nn.Linear(hidden_dim, vocab_size)
        self.fc_bounds = nn.Linear(hidden_dim, 4)
        self.dropout = nn.Dropout(0.2)
    def forward(self, x, hidden=None):
        x = self.embed(x)
        x = self.dropout(x)
        out, hidden = self.lstm(x, hidden)
        token_logits = self.fc_token(out)
        bounds = torch.sigmoid(self.fc_bounds(out))
        return token_logits, bounds, hidden
    def sample(self, batch_size, start_token, end_token, pad_token, max_len, device):
        self.eval()
        samples = []
        layouts = []
        with torch.no_grad():
            for _ in range(batch_size):
                seq = [start_token]
                layout = [[0, 0, 0, 0]]
                input_token = torch.tensor([[start_token]], device=device)
                hidden = None
                for _ in range(max_len - 2):
                    token_logits, bounds, hidden = self.forward(input_token, hidden)
                    token_probs = F.softmax(token_logits[0, -1], dim=-1)
                    next_token = torch.multinomial(token_probs, 1).item()
                    seq.append(next_token)
                    layout.append(bounds[0, -1].tolist())
                    if next_token == end_token:
                        break
                    input_token = torch.tensor([[next_token]], device=device)
                if seq[-1] != end_token:
                    seq.append(end_token)
                    layout.append([0, 0, 0, 0])
                while len(seq) < max_len:
                    seq.append(pad_token)
                    layout.append([0, 0, 0, 0])
                samples.append(seq[:max_len])
                layouts.append(layout[:max_len])
        self.train()
        return {
            'sequence': torch.tensor(samples, device=device, dtype=torch.long),
            'layout': torch.tensor(layouts, device=device, dtype=torch.float)
        }

class LayoutDiscriminator(nn.Module):
    def __init__(self, vocab_size, embed_dim=256, hidden_dim=256, num_layers=2):
        super().__init__()
        self.embed = nn.Embedding(vocab_size, embed_dim)
        self.lstm = nn.LSTM(embed_dim + 4, hidden_dim, num_layers, batch_first=True, bidirectional=True)
        self.classifier = nn.utils.spectral_norm(nn.Linear(hidden_dim * 2, 1))
        self.dropout = nn.Dropout(0.3)
    def forward(self, x):
        seq_embed = self.embed(x['sequence'])
        combined = torch.cat([seq_embed, x['layout']], dim=-1)
        combined = combined + 0.05 * torch.randn_like(combined)
        combined = self.dropout(combined)
        _, (hidden, _) = self.lstm(combined)
        features = torch.cat([hidden[-2], hidden[-1]], dim=1)
        return self.classifier(features)

# === STYLE & STRUCTURE LOSSES ===
class StyleHomogeneityLoss(nn.Module):
    def __init__(self):
        super().__init__()
    def forward(self, layouts):
        batch_size = layouts.size(0)
        loss = 0.0
        for i in range(batch_size):
            valid_mask = (layouts[i].sum(dim=-1) != 0)
            if valid_mask.sum() <= 1:
                continue
            valid_layouts = layouts[i][valid_mask]
            widths = valid_layouts[:, 2]
            heights = valid_layouts[:, 3]
            valid_size_mask = (heights > 1e-6) & (widths > 1e-6)
            if valid_size_mask.sum() <= 1:
                continue
            valid_widths = widths[valid_size_mask]
            valid_heights = heights[valid_size_mask]
            wh_ratios = valid_widths / valid_heights
            ratio_variance = torch.var(wh_ratios)
            x_coords = valid_layouts[:, 0]
            y_coords = valid_layouts[:, 1]
            x_alignment = 0
            y_alignment = 0
            n_elements = valid_layouts.size(0)
            for j in range(n_elements):
                x_aligned = torch.sum((torch.abs(x_coords - x_coords[j]) < 0.1).float())
                y_aligned = torch.sum((torch.abs(y_coords - y_coords[j]) < 0.1).float())
                x_alignment += x_aligned
                y_alignment += y_aligned
            x_alignment = x_alignment / (n_elements * n_elements)
            y_alignment = y_alignment / (n_elements * n_elements)
            element_loss = ratio_variance + 0.1 * (1 - x_alignment) + 0.1 * (1 - y_alignment)
            loss += element_loss
        if batch_size > 0:
            print(f"[DEBUG] Style loss: {loss / batch_size:.4f}, layout mean: {layouts.mean().item():.4f}")
        return loss / batch_size if batch_size > 0 else torch.tensor(0.0, device=layouts.device)

class StructureConsistencyLoss(nn.Module):
    def __init__(self):
        super().__init__()
    def forward(self, pred_sequences, target_sequences):
        batch_size = pred_sequences.size(0)
        loss = 0.0
        for i in range(batch_size):
            pred_valid = pred_sequences[i][pred_sequences[i] != 0]
            target_valid = target_sequences[i][target_sequences[i] != 0]
            if len(pred_valid) == 0 or len(target_valid) == 0:
                continue
            m, n = len(pred_valid), len(target_valid)
            dp = torch.zeros((m+1, n+1), device=pred_sequences.device)
            for j in range(n+1):
                dp[0, j] = j
            for i2 in range(m+1):
                dp[i2, 0] = i2
            for i2 in range(1, m+1):
                for j2 in range(1, n+1):
                    if pred_valid[i2-1] == target_valid[j2-1]:
                        dp[i2, j2] = dp[i2-1, j2-1]
                    else:
                        dp[i2, j2] = 1 + torch.min(torch.stack([
                            dp[i2-1, j2], dp[i2, j2-1], dp[i2-1, j2-1]
                        ]))
            normalized_distance = dp[m, n] / max(m, n, 1)
            loss += normalized_distance
        return loss / batch_size if batch_size > 0 else torch.tensor(0.0, device=pred_sequences.device)

# === CHECKPOINT UTILITIES ===
def save_checkpoint(epoch, generator, discriminator, g_optimizer, d_optimizer, vocab, loss_history):
    checkpoint_path = CHECKPOINT_DIR / f'phase2_epoch_{epoch+1}.pt'
    checkpoint = {
        'epoch': epoch,
        'generator_state_dict': generator.state_dict(),
        'discriminator_state_dict': discriminator.state_dict(),
        'g_optimizer_state_dict': g_optimizer.state_dict(),
        'd_optimizer_state_dict': d_optimizer.state_dict(),
        'vocab': vocab,
        'loss_history': loss_history
    }
    torch.save(checkpoint, checkpoint_path)
    print(f"✅ Checkpoint saved: {checkpoint_path}")
    return checkpoint_path

def load_latest_checkpoint(generator, discriminator):
    checkpoints = sorted(CHECKPOINT_DIR.glob('phase2_epoch_*.pt'))
    if not checkpoints:
        print("No checkpoints found. Starting from scratch.")
        return 0, [], None
    latest_checkpoint = checkpoints[-1]
    print(f"🔄 Loading checkpoint: {latest_checkpoint}")
    checkpoint = torch.load(latest_checkpoint, map_location=DEVICE)
    generator.load_state_dict(checkpoint['generator_state_dict'], strict=False)
    discriminator.load_state_dict(checkpoint['discriminator_state_dict'])
    start_epoch = checkpoint['epoch'] + 1
    loss_history = checkpoint.get('loss_history', [])
    vocab = checkpoint['vocab']
    print(f"✅ Resumed from epoch {checkpoint['epoch']}")
    return start_epoch, loss_history, vocab

# === MAIN TRAINING FUNCTION ===
def train_phase2():
    print("Preparing Phase 2 data...")
    prepare_phase2_data()
    dataset = GUILayoutDataset(
        os.path.join(PHASE2_DATA_PATH, "gui_metadata.json"),
        os.path.join(PHASE2_DATA_PATH, "vocab.json")
    )
    loader = DataLoader(
        dataset,
        batch_size=BATCH_SIZE,
        shuffle=True,
        num_workers=NUM_WORKERS,
        pin_memory=PIN_MEMORY
    )
    with open(os.path.join(PHASE2_DATA_PATH, "vocab.json")) as f:
        vocab = json.load(f)
    vocab_size = len(vocab)
    start_token = vocab['<START>']
    end_token = vocab['<END>']
    pad_token = vocab['<PAD>']
    print(f"Dataset size: {len(dataset)} GUI layouts")
    print(f"Vocabulary size: {vocab_size}")

    generator = LayoutGenerator(vocab_size).to(DEVICE)
    discriminator = LayoutDiscriminator(vocab_size).to(DEVICE)
    g_optimizer = optim.Adam(generator.parameters(), lr=G_LR, betas=(0.5, 0.999))
    d_optimizer = optim.Adam(discriminator.parameters(), lr=D_LR, betas=(0.5, 0.999))
    adv_criterion = nn.BCEWithLogitsLoss()
    style_criterion = StyleHomogeneityLoss()
    struct_criterion = StructureConsistencyLoss()

    start_epoch, loss_history, checkpoint_vocab = load_latest_checkpoint(
        generator, discriminator
    )
    if checkpoint_vocab:
        vocab = checkpoint_vocab
    print(f"🚀 Starting Phase 2 training from epoch {start_epoch}")

    try:
        for epoch in range(start_epoch, EPOCHS):
            g_losses, d_losses = [], []
            style_losses, struct_losses = [], []
            for batch_idx, batch in enumerate(tqdm(loader, desc=f"Epoch {epoch+1}/{EPOCHS}")):
                real_seq = batch['sequence'].to(DEVICE)
                real_layout = batch['layout'].to(DEVICE)
                real_data = {'sequence': real_seq, 'layout': real_layout}
                batch_size = real_seq.size(0)

                # === Discriminator update ===
                for _ in range(D_STEPS):
                    d_optimizer.zero_grad()
                    fake_data = generator.sample(batch_size, start_token, end_token, pad_token, MAX_COMPONENTS, DEVICE)
                    real_pred = discriminator(real_data)
                    fake_pred = discriminator(fake_data)
                    real_labels = torch.ones_like(real_pred) * (0.9 + 0.05 * torch.rand_like(real_pred))
                    fake_labels = torch.zeros_like(fake_pred) + 0.05 * torch.rand_like(fake_pred)
                    d_loss_real = adv_criterion(real_pred, real_labels)
                    d_loss_fake = adv_criterion(fake_pred, fake_labels)
                    d_loss = (d_loss_real + d_loss_fake) / 2
                    d_loss.backward()
                    d_optimizer.step()
                    d_losses.append(d_loss.item())

                # === Generator update ===
                for _ in range(G_STEPS):
                    g_optimizer.zero_grad()
                    fake_data = generator.sample(batch_size, start_token, end_token, pad_token, MAX_COMPONENTS, DEVICE)
                    fake_pred = discriminator(fake_data)
                    adv_loss = adv_criterion(fake_pred, torch.ones_like(fake_pred))
                    style_loss = style_criterion(fake_data['layout'])
                    struct_loss = struct_criterion(fake_data['sequence'], real_data['sequence'])
                    g_loss = 0.7 * adv_loss + 0.2 * style_loss + 0.1 * struct_loss
                    if epoch < 5:
                        logits, _, _ = generator(real_seq[:, :-1])
                        teacher_loss = F.cross_entropy(logits.reshape(-1, vocab_size), real_seq[:, 1:].reshape(-1), ignore_index=pad_token)
                        g_loss = g_loss + 0.3 * teacher_loss
                    g_loss.backward()
                    g_optimizer.step()
                    g_losses.append(g_loss.item())
                    style_losses.append(style_loss.item())
                    struct_losses.append(struct_loss.item())

                if (batch_idx + 1) % PRINT_FREQ == 0:
                    print(f"Epoch {epoch+1}, Batch {batch_idx+1}/{len(loader)}: "
                          f"D Loss={d_loss.item():.4f}, G Loss={g_loss.item():.4f}, "
                          f"Style Loss={style_loss.item():.4f}, Struct Loss={struct_loss.item():.4f}")

            avg_d_loss = sum(d_losses) / len(d_losses) if d_losses else 0
            avg_g_loss = sum(g_losses) / len(g_losses) if g_losses else 0
            avg_style_loss = sum(style_losses) / len(style_losses) if style_losses else 0
            avg_struct_loss = sum(struct_losses) / len(struct_losses) if struct_losses else 0
            loss_history.append({
                'epoch': epoch,
                'd_loss': avg_d_loss,
                'g_loss': avg_g_loss,
                'style_loss': avg_style_loss,
                'struct_loss': avg_struct_loss
            })
            print(f"Epoch {epoch+1}/{EPOCHS} Summary:")
            print(f"  D Loss: {avg_d_loss:.4f}")
            print(f"  G Loss: {avg_g_loss:.4f}")
            print(f"  Style Loss: {avg_style_loss:.4f}")
            print(f"  Structure Loss: {avg_struct_loss:.4f}")
            with torch.no_grad():
                sample_data = generator.sample(1, start_token, end_token, pad_token, MAX_COMPONENTS, DEVICE)
                sample_seq = sample_data['sequence'][0].cpu().tolist()
                sample_layout = sample_data['layout'][0].cpu().tolist()
                inv_vocab = {v: k for k, v in vocab.items()}
                sample_components = [inv_vocab.get(token, '<UNK>') for token in sample_seq if token not in [start_token, end_token, pad_token]]
                print(f"Sample layout: {sample_components[:5]}")
                with open(SAMPLE_OUTPUT_PATH, "a") as f:
                    f.write(json.dumps({
                        "epoch": epoch+1,
                        "components": sample_components,
                        "layout": sample_layout
                    }) + "\n")
            save_checkpoint(epoch, generator, discriminator, g_optimizer, d_optimizer, vocab, loss_history)
        print("🎉 Phase 2 training completed!")
    except KeyboardInterrupt:
        print("\n⚠️ Training interrupted by user")
        save_checkpoint(epoch, generator, discriminator, g_optimizer, d_optimizer, vocab, loss_history)
        print("🚨 Emergency checkpoint saved!")
    except Exception as e:
        print(f"❌ Training error: {e}")
        try:
            save_checkpoint(epoch, generator, discriminator, g_optimizer, d_optimizer, vocab, loss_history)
            print("🚨 Emergency checkpoint saved!")
        except:
            print("❌ Could not save emergency checkpoint")

if __name__ == "__main__":
    train_phase2()
```

```python
import torch
import time
import matplotlib.pyplot as plt

# Dummy matrix size to simulate workload
size = 2048
repeats = 50

# Function to time a matrix operation
def benchmark(device):
    x = torch.randn(size, size).to(device)
    y = torch.randn(size, size).to(device)
    torch.cuda.synchronize() if device.type == 'cuda' else None

    start_time = time.time()
    for _ in range(repeats):
        z = torch.matmul(x, y)
    torch.cuda.synchronize() if device.type == 'cuda' else None
    end_time = time.time()

    return end_time - start_time

# Run on CPU
cpu_time = benchmark(torch.device("cpu"))
print(f"CPU time: {cpu_time:.2f} seconds")

# Run on GPU
gpu_time = benchmark(torch.device("cuda"))
print(f"GPU time: {gpu_time:.2f} seconds")

# Plot comparison
plt.bar(["CPU", "GPU"], [cpu_time, gpu_time], color=["red", "green"])
plt.ylabel("Time (s)")
plt.title("Matrix Multiplication Time Comparison")
plt.show()
```

```python
# Simplified architecture snippet
class Generator(nn.Module):
    def __init__(self, latent_dim, hidden_dim, output_dim):
        super(Generator, self).__init__()
        self.lstm = nn.LSTM(latent_dim, hidden_dim, num_layers=2, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, z):
        out, _ = self.lstm(z)
        layout = self.fc(out)
        return layout
```

```python
class Discriminator(nn.Module):
    def __init__(self, input_dim, hidden_dim):
        super(Discriminator, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.LeakyReLU(0.2),
            nn.Linear(hidden_dim, 1),
            nn.Sigmoid()
        )

    def forward(self, layout_sequence):
        validity = self.model(layout_sequence)
        return validity
```
